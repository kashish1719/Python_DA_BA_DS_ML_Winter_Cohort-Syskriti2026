```python
## Simple Calculator
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

sum = num1 + num2
difference = abs(num1 - num2)
product = num1 * num2
quotient = num1 / num2
remainder = num1 % num2 

print("Sum", sum)
print("Difference", difference)
print("Product", product)
print("Quotient", quotient)
print("Remainder", remainder)
```
