```python
## MEAN
data = [4, 2, 7, 2, 9, 4, 2]
sum = 0
for num in data:
    sum = sum + num
mean = sum / len(data)
print("Mean of data: ", mean)
```

```python
## MEDIAN
data = [4, 12, 7, 2, 9, 4, 5]
# sorted_data = []
n = len(data)
# sort using bubble sort -- O(n^2)
for i in range(n):
    for j in range(0, n - i - 1):
        if data[j] > data[j + 1]:
            temp = data[j]
            data[j] = data[j + 1]
            data[j + 1] = temp

if(n % 2 != 0):
    median = data[n // 2]
else:
    median = (((n // 2) - 1) + (n // 2)) / 2

print("Median of data: ", median)
```

```python
## MODE
data = [4, 2, 7, 2, 9, 4, 2]
max_count = 0
mode = None
for i in data:
    count = 0
    for j in data:
        if i == j:
            count += 1
    if count > max_count:
        max_count = count
        mode = i

print("Mode ", mode)
print(f"Frequency of {mode} is {max_count}")
```
